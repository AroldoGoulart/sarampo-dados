```python

import pandas as pd
import numpy as np
import math

# total de vacinados
x = np.array([4837858, 5106270, 4161884, 3458574])
# infectados
y = np.array([10346, 20901, 8448, 668])

# array de interceisao
sub =  pd.array([])

n = x.size

# formula
b = (n*np.sum(x*y)-np.sum(x)*np.sum(y))/(n*np.sum(x**2)-np.sum(x)**2)
a = y.mean() -b *x.mean()
sub = a + b*x

#periodo de datas
dates = pd.date_range("2018", periods=4, freq='A')


# formula, total, infectados, respectivamente
g = pd.Series((sub), index=dates)
gt = pd.Series((x), index=dates)
gi = pd.Series((y), index=dates)

g.plot(figsize=(16,9))
gi.plot()
gt.plot()
```
